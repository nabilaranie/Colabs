# Task 01 : Apakah ada dosen yang cenderung memberikan nilai cenderung lebih tinggi (atau lebih rendah) dibandingkan dosen lainnya?

Diskusi 01:
* Apakah baik menilai (judge) hanya dari rata-rata.
* Rata-rata + Variansi cukup?
* Critical Thinking: Hati-hati membandingkan mata pelajarannya yang berbeda, bisa jadi misal suatu mata kuliah memang lebih sulit dari MK lain. Atau MK dasar VS MK di semester akhir.
* Lalu apa yang harus dilakukan untuk melakukan perbandingan yang lebih baik? ==> standarisasi soal, membandingkan MK sama di tempat berbeda (misal MK pelayanan).
* Data Science perlu di barengi dengan business understanding dan terkadang studi kualitatif untuk mendapatkan insight yang baik.

```python
#Mencari rata-rata dan standar deviasi pada masing-masing dosen
nilai_dosen = nilai.groupby(nilai['NIDN']).mean()[['nilai']]
nilai_dosen = nilai_dosen.rename({'nilai': 'mean_nilai'}, axis=1)
nilai_dosen = pd.merge(dosen, nilai_dosen, left_on='NIDN', right_index=True)
nilai_dosen = nilai_dosen[['NIDN', 'nama', 'mean_nilai']]
sb = nilai.groupby(nilai['NIDN']).std()[['nilai']]
sb = sb.rename({'nilai': 'std_nilai'}, axis=1)
nilai_dosen = nilai_dosen.set_index(['NIDN'])
nilai_dosen = pd.concat([nilai_dosen, sb], axis=1)
nilai_dosen
```

```python
#Hasil dengan membandingkan MK yang sama (Kalkulus Dasar)
kalk = nilai[nilai['id_mata_kuliah'] == 'KAL01']
sdkalk = kalk.groupby(kalk['NIDN']).std()[['nilai']]
sdkalk = sdkalk.rename({'nilai': 'std_nilai'}, axis=1)
kalk = kalk.groupby(kalk['NIDN']).mean()[['nilai']]
nama_dosen = dosen.set_index(['NIDN']).drop(['email', 'Pendidikan_Terakhir', 'Telephone', 'Jenis_Kelamin'], axis=1)
kalk = pd.concat([nama_dosen, kalk, sdkalk], axis=1).dropna()
kalk = kalk.rename({'nilai' : 'mean_nilai'}, axis=1)
kalk
```

```python
#Hasil dengan membandingkan MK yang sama (Kalkulus Lanjut)
kalk = nilai[nilai['id_mata_kuliah'] == 'KAL02']
sdkalk = kalk.groupby(kalk['NIDN']).std()[['nilai']]
sdkalk = sdkalk.rename({'nilai': 'std_nilai'}, axis=1)
kalk = kalk.groupby(kalk['NIDN']).mean()[['nilai']]
nama_dosen = dosen.set_index(['NIDN']).drop(['email', 'Pendidikan_Terakhir', 'Telephone', 'Jenis_Kelamin'], axis=1)
kalk = pd.concat([nama_dosen, kalk, sdkalk], axis=1).dropna()
kalk = kalk.rename({'nilai' : 'mean_nilai'}, axis=1)
kalk
```

```python
#Hasil dengan membandingkan MK yang sama (Ilmu Komputer)
kalk = nilai[nilai['id_mata_kuliah'] == 'CS01']
sdkalk = kalk.groupby(kalk['NIDN']).std()[['nilai']]
sdkalk = sdkalk.rename({'nilai': 'std_nilai'}, axis=1)
kalk = kalk.groupby(kalk['NIDN']).mean()[['nilai']]
nama_dosen = dosen.set_index(['NIDN']).drop(['email', 'Pendidikan_Terakhir', 'Telephone', 'Jenis_Kelamin'], axis=1)
kalk = pd.concat([nama_dosen, kalk, sdkalk], axis=1).dropna()
kalk = kalk.rename({'nilai' : 'mean_nilai'}, axis=1)
kalk
```

# Task 03 : Apakah terdapat perbedaan performa akademik antara mahasiswa yang berasal dari Madrasah, SMK, dan SMU?


```python
msw = pd.concat([nilai.set_index(['nim']), mahasiswa.set_index(['nim'])], axis=1)
sdmsw = msw.groupby(msw['Asal_Sekolah']).std()[['nilai']]
sdmsw = sdmsw.rename({'nilai': 'std_nilai'}, axis=1)
ctmsw = msw.groupby(msw['Asal_Sekolah']).count()['Asal_Daerah']
msw = msw.groupby(msw['Asal_Sekolah']).mean()[['nilai']]
msw = pd.concat([msw, sdmsw, ctmsw], axis=1).rename({'Asal_Daerah': 'jumlah'})
msw = msw.rename({'nilai' : 'mean_nilai'}, axis=1)
msw
```

```python
#Hasil dari MK yang sama
msw = pd.concat([nilai.set_index(['nim']), mahasiswa.set_index(['nim'])], axis=1)
msw['MK_ASAL'] = msw['id_mata_kuliah'] + '_' + msw['Asal_Sekolah']
#msw
sdmsw = msw.groupby(msw['MK_ASAL']).std()[['nilai']]
sdmsw = sdmsw.rename({'nilai': 'std_nilai'}, axis=1)
ctmsw = msw.groupby(msw['MK_ASAL']).count()['Asal_Daerah']
msw = msw.groupby(msw['MK_ASAL']).mean()[['nilai']]
msw = pd.concat([msw, sdmsw, ctmsw], axis=1).rename({'Asal_Daerah' : 'jumlah'}, axis=1)
msw = msw.rename({'nilai' : 'mean_nilai'}, axis=1)
msw
```

# Task 04 : Apakah ada perbedaan yang nampak di data terhadap kemampuan akademik antara mahasiswa dan mahasiswi?

```python
msw = pd.concat([nilai.set_index(['nim']), mahasiswa.set_index(['nim'])], axis=1)
sdmsw = msw.groupby(msw['Gender']).std()[['nilai']]
sdmsw = sdmsw.rename({'nilai': 'std_nilai'}, axis=1)
msw = msw.groupby(msw['Gender']).mean()[['nilai']]
msw = pd.concat([msw, sdmsw], axis=1)
msw = msw.rename({'nilai' : 'mean_nilai'}, axis=1)
msw
```

```python
#Hasil dari MK yang sama
msw = pd.concat([nilai.set_index(['nim']), mahasiswa.set_index(['nim'])], axis=1)
msw['MK_GENDER'] = msw['id_mata_kuliah'] + '_' + msw['Gender']
#msw
sdmsw = msw.groupby(msw['MK_GENDER']).std()[['nilai']]
sdmsw = sdmsw.rename({'nilai': 'std_nilai'}, axis=1)
msw = msw.groupby(msw['MK_GENDER']).mean()[['nilai']]
msw = pd.concat([msw, sdmsw], axis=1)
msw = msw.rename({'nilai' : 'mean_nilai'}, axis=1)
msw
```
