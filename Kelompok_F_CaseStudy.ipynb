# Soal 01: Multiplicative Congruential

1. Buatlah sebuah fungsi pembangkit bilangan acak dengan fungsi Greenberger (Multiplicative congruential.) dengan ketentuan modifikasi sebagai berikut:
 * 	Xo = 11, a=7, c=5, m = $2^{32}$
 * 	Input fungsi adalah $N$ (jumlah bilangan yang akan dibangkitkan)
 *	Output fungsi adalah sebuah skalar: rata-rata semua $N$ suku dibarisan yang dibangkitkan.
 *  Jangan lupa untuk menormalisasi semua sukunya dengan $m$.

2. Jika $X = x_1, x_2, ..., x_N$ adalah barisan $N$ output dari fungsi diatas, apakah distribusi dari $X$? Jelaskan.

3. Apakah $X$ memiliki periode? jelaskan.

```python
# Jawaban Soal 01 dapat dituliskan di cell ini.
# Tambahkan cell lain jika diperlukan

def Greenberger(xo,m,a,c):
  return(a*xo+c)%m

m=2**32
a=7
c=5
xo=11
x1=Greenberger(xo,m,a,c)
result = []
result2 = []
total = 0
total2 = 0
for i in range(1000):
    xo = Greenberger(xo,m, a, c)
    print(xo, end = ', ')
    result.append(xo)
    result2.append(xo/m)
    total += xo
    total2 += xo/m
print(f"X = {', '.join(map(str, result))}")
print(f"Rata-rata = {total/N}")
print(f"X_normalisasi = {', '.join(map(str, result2))}")
print(f"Rata-rata_normalisasi= {total2/N}")
```

```python
#Untuk melihat apakah ada pengulangan maka menggunakan while loop
result = []
total = 0
run = True
while run:
    xo = Greenberger(xo, m, a, c)
    if xo not in result:
        run = True
    else:
        run = False
    result.append(xo)
    total += xo
```

Ternyata kode program di atas tidak akan pernah berhenti (selalu melakukan looping). Jadi dapat disimpulkan X tidak memiliki periode pengulangan

```python
#Jika tidak menggunakan Pure Python Module

def generate_poisson(lambda_parameter):
    i = 0
    p = math.exp(-lambda_parameter)
    F = p
    U = random.random()

    while True:
        if U < F:
            return i
        else:
            i += 1
            p = (1.0 * p * lambda_parameter) / i
            F = F + p
            U = random.random()

lambda_parameter = 5
sample_size = 100000

poisson_data = [generate_poisson(lambda_parameter) for _ in range(sample_size)]

plt.hist(poisson_data, bins=10, density=True, alpha=0.6, color='g')
plt.xlabel('Nilai Poisson')
plt.ylabel('Frekuensi Normalisasi')
plt.title('Histogram Distribusi Poisson')
plt.show()
```

```python
#Jawaban alternatif

def Poisson(i, p, F, U):
  if U < F:
    return i
  else:
    #print(U, F, p)
    p = 1*p/(i + 1)
    F = F + p
    i += 1
    return Poisson(i, p, F, U)

seed(1)
random = rand(100000)
result = {}
for U in random:
  temp = Poisson(0, np.exp(-1), np.exp(-1), U)
  if temp not in result:
    result[temp] = 1
  else:
    result[temp] += 1

fixedresult = []
for i in result:
  fixedresult.append([i, result[i]])
fixedresult.sort(key=lambda x:x[0], reverse=False)

randomX = []
randomY = []
for i in fixedresult:
  randomX.append(i[0])
  randomY.append(i[1])

temp = np.random.poisson(1, 100000)
poissonresult = {}
for i in temp:
  if i not in poissonresult:
    poissonresult[i] = 1
  else:
    poissonresult[i] += 1

fixedpoissonresult = []
for i in poissonresult:
  fixedpoissonresult.append([i, poissonresult[i]])
fixedpoissonresult.sort(key=lambda x:x[0], reverse=False)

poissonX = []
poissonY = []
for i in fixedpoissonresult:
  poissonX.append(i[0])
  poissonY.append(i[1])
```

```python
plt.plot(randomX, randomY, color='blue', label='Metode Invers Distribusi', linestyle='dashed')
plt.plot(poissonX, poissonY, color='red', label='Hasil Langsung Dari NumPy', linestyle='dashed')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', markerscale=5.0)
#plt.xlabel('Nilai X')
plt.ylabel('Frekuensi')
plt.title('Plot Distribusi Poisson')
```

# Metode Crude Monte Carlo dengan N = 100







```python
#Metode Crude

def f(x):
  return (np.exp(x)-1)/(np.exp(1)-1)

seed(55)

N=100
teta = 0.418023
x_min, x_max = (0,1)
randx = []
for i in range(N):
  randx.append(np.random.uniform(x_min, x_max))

print('$$$ Crude Monte Carlo $$$')
y = f(randx)
sum_of_samples = 0
for i in y :
  sum_of_samples += i

crude_monte_carlo = (x_max - x_min) * float(sum_of_samples/N)

print('Hasil Approksimasi = ', crude_monte_carlo)
print('Error = ', abs(teta-crude_monte_carlo)) #Perbandingan solusi eksak dengan Crude Monte Carlo
print('Standar Error = ', np.sqrt(teta*(1-teta)/N))
print('Aproksimasi Error = ', np.sqrt(crude_monte_carlo*(1-crude_monte_carlo)/N))
print('Confidence Interval 95% = ', crude_monte_carlo, ' +- ', np.sqrt(crude_monte_carlo*(1-crude_monte_carlo)/N))
```

```python
#Hit or Miss
def f(x):
    return (np.exp(x)-1)/(np.exp(1)-1)

seed(55)

N=100
teta = 0.418023
#sigma = 0.286 #Solusi Eksak teta dan sigma
xrand = rand(N)
yrand = rand(N) # Random Uniform x dan
print('$$$ Hit or Miss Monte Carlo $$$') # Hit or Miss Monte Carlo

fxrand=f(xrand)
CheckValue = fxrand>=yrand
Hit_or_Miss=sum(CheckValue)/N

print('Hasil Approksimasi = ', Hit_or_Miss)
print('Error = ', abs(teta-Hit_or_Miss))
print('Standar Error = ', np.sqrt(teta*(1-teta)/N))
print('Aproksimasi Error = ', np.sqrt(Hit_or_Miss*(1-Hit_or_Miss)/N))
print('Confidence Interval 95% = ', Hit_or_Miss, ' +- ', np.sqrt(Hit_or_Miss*(1-Hit_or_Miss)/N))
```

```python
print("Perbandingan antara Crude Monte Carlo dan Hit or Miss")
print(f"Residual = {abs(Hit_or_Miss-crude_monte_carlo)}")
```
